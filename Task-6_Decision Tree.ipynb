## Task 6 : Prediction using Decision Tree Algorithm

#### _Objective: Create the Decision Tree classifier and visualize it graphically. The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly._

### 1. Importing the dataset and all required libraries

```python
import pandas as pd
import numpy as np
import sklearn.datasets as datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from matplotlib import pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
#Loading the iris dataset
irs=datasets.load_iris()

#Forming the iris dataframe
dataframe=pd.DataFrame(irs.data, columns=irs.feature_names)
```

```python
#Read the dataset
dataframe.head()
```

### 2. Prepare the data for the Decision Tree model

```python
#Preparing the data
x=irs.data
y=irs.target


#spliting the dataset into 20% test data and 80% train data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=1)
print("x train data", x_train.shape)
print("x test data", x_test.shape)
print("y train data", y_train.shape)
print("y test data", y_test.shape)
```

+ **Here 120 data for training the model and 30 data for testing the model.**

### 3. Create the Decision Tree model and train the model

```python
#Creating Decision Tree classifier
dtree=DecisionTreeClassifier()

#Train Decision Tree Classifer
dtree.fit(x_train,y_train)

print("Decision Tree Classifer Created Successfully")
```

### 4. Predict the Test Result

```python
#Predict the response for test dataset
y_predict=dtree.predict(x_test)
y_predict
```

### 5. Evaluating the Model

```python
score=accuracy_score(y_predict,y_test)
print(f"Accuracy score is {score} i.e. {score*100}%")
```

### 6. Visualizing Decision Tree

```python
plt.figure(figsize=(20,15))
tree.plot_tree(dtree, filled=True, rounded=True, feature_names=irs.feature_names, class_names=irs.target_names)
plt.show()
```
